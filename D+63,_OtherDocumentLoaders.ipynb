```python
#UnstructuredPowerPointLoader 불러오기
from langchain_community.document_loaders import UnstructuredPowerPointLoader

#mode=elements를 통해 pptx의 요소별로 Document 객체로 가져오기
loader = UnstructuredPowerPointLoader(r"/content/drive/MyDrive/WB/2025JPO/Reading/Ageing-and-Longevity-Seminar-Series-Anita.pptx", mode="elements")

#pptx 파일을 분할 로드하기
data = loader.load_and_split()

data[3]
```

```python
for i in data:
    if i.metadata['page_number'] == 3:
        print(i.metadata['category'])
        print(i.page_content)
        print("\n")
```

## **인터넷 정보 로드하기, WebBaseLoader**

```python
from langchain_community.document_loaders import WebBaseLoader
#텍스트 추출할 URL 입력
loader = WebBaseLoader("https://openknowledge.worldbank.org/server/api/core/bitstreams/99905eba-11a6-43ac-96bb-29cd8284f6ea/content")
#ssl verification 에러 방지를 위한 코드
loader.requests_kwargs = {'verify':False}
data = loader.load()
data
#웹 페이지의 텍스트 추출하기
```

**[Headline만 가져오기]**

```python
import bs4
from langchain_community.document_loaders import WebBaseLoader
#텍스트 추출할 URL 입력
loader = WebBaseLoader("https://openknowledge.worldbank.org/server/api/core/bitstreams/99905eba-11a6-43ac-96bb-29cd8284f6ea/content",
                        bs_kwargs=dict(
                            parse_only=bs4.SoupStrainer(
                                class_=("headlineStack top-headlines")
                                                        )
                                        )
                      )
#ssl verification 에러 방지를 위한 코드
loader.requests_kwargs = {'verify':False}
data = loader.load()
data
```

```python
# 여려개의 웹 페이지에서 텍스트를 동시에 추출하기
loader = WebBaseLoader(["https://openknowledge.worldbank.org/entities/publication/6474c5c1-8124-4d13-8f0d-dcb098d274c6", "https://openknowledge.worldbank.org/entities/publication/0445391c-9b64-4e6e-ad99-617dce4f27bc"])
docs = loader.load()
docs
```

```python
loader = WebBaseLoader(["https://www.espn.com/", "https://www.google.com/"])
docs = loader.load()
docs
```

## **특정 경로 내의 모든 파일 불러오기, DirectoryLoader**

```python
!pip install unstructured[pdf]
```

```python
from langchain_community.document_loaders import DirectoryLoader
#첫번째 매개변수로 경로 입력, glob에 해당 경로에서 불러들일 파일의 형식 지정
#*는 모든 문자를 표현하는 와일드카드로, .pdf로 끝나는 모든 파일을 의미함
#디렉토리에 있는 파일만 가능한 경우임
loader = DirectoryLoader(r"/content/drive/MyDrive/WB/2025JPO/Reading/", glob="*.pdf")
docs = loader.load()
[i.metadata['source'] for i in docs]

```
